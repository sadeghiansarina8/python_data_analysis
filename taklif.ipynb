```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# لیست شهرها
cities = [
    {"name": "Tehran", "lat": 35.6892, "lon": 51.3890, "pop": 8.7},
    {"name": "Mashhad", "lat": 36.2988, "lon": 59.6062, "pop": 3.2},
    {"name": "Isfahan", "lat": 32.6525, "lon": 51.6776, "pop": 2.0},
    {"name": "Shiraz", "lat": 29.5918, "lon": 52.5836, "pop": 1.8},
    {"name": "Tabriz", "lat": 38.0962, "lon": 46.2738, "pop": 1.6},
    {"name": "Qom", "lat": 34.6401, "lon": 50.8764, "pop": 1.5},
    {"name": "Ahvaz", "lat": 31.3183, "lon": 48.6706, "pop": 1.4},
    {"name": "Kermanshah", "lat": 34.3142, "lon": 47.0650, "pop": 1.1},
    {"name": "Urmia", "lat": 37.5552, "lon": 45.0720, "pop": 0.7},
    {"name": "Yazd", "lat": 31.8974, "lon": 54.3569, "pop": 0.6}
]

# رنگ‌ها برای هر شهر
colors = ['red', 'blue', 'green', 'purple', 'orange', 'cyan', 'magenta', 'yellow', 'brown', 'pink']

# ایجاد شکل و محور
fig = plt.figure(figsize=(12,10))
ax = plt.axes(projection=ccrs.Mercator())

# محدوده نقشه
ax.set_extent([44, 63, 25, 40], crs=ccrs.PlateCarree())

# خشکی و دریا
ax.add_feature(cfeature.LAND, facecolor='gray')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')
ax.add_feature(cfeature.STATES)

# مرزهای کشورها
borders = cfeature.NaturalEarthFeature(
    category='cultural',
    name='admin_0_countries',
    scale='50m',
    facecolor='none',
    edgecolor='black',
    )

ax.add_feature(borders, edgecolor='black', linewidth=0.5)  # مرز کل کشورها


# رسم شهرها
for i, city in enumerate(cities):
    size = city["pop"] * 300
    ax.scatter(city["lon"], city["lat"], s=size, color=colors[i], alpha=0.5, transform=ccrs.PlateCarree())
    ax.text(city["lon"] + 0.3, city["lat"] + 0.3, city["name"], transform=ccrs.PlateCarree())

# عنوان
ax.set_title("Iranian 10 City Population")

plt.show()
```
